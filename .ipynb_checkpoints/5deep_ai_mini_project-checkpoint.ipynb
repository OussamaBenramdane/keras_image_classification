# Filtrer les images corrompues

**Cette fonction nous permet de filtrer les images corrompus après avoir parcouru en boucle le dossier et les sous-dossiers.
Ensuite vérifier si le JFIF est inclus dans l'en-tête des métadonnées de l'image.**


```python
num_skipped = 0
for folder_name in ("buildings", "forest","glacier","mountain","sea","street","corrupted"):
    folder_path = os.path.join("seg_train", folder_name)
    for fname in os.listdir(folder_path):
        fpath = os.path.join(folder_path, fname)
        try:
            fobj = open(fpath, "rb")
            is_jfif = tf.compat.as_bytes("JFIF") in fobj.peek(10)
        finally:
            fobj.close()

        if not is_jfif:
            num_skipped += 1
            # Delete corrupted image
            os.remove(fpath)

print("Deleted %d images" % num_skipped)
```

# Définitions des ensembles de données

**Ici, nous chargerons les différents jeux de données que nous allons utiliser à partir du dépôt d'images pour obtenir les jeux de validation et d'entraînement.**


```python
#dimenssion
image_size = (150, 150)
batch_size = 32

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "seg_train",
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=image_size,# redimensionnement des images
    batch_size=batch_size,
)
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "seg_train",
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=image_size,
    batch_size=batch_size,
)
```

**Montrer les classes de l'ensemble de données**

```python
class_names = train_ds.class_names
print(class_names)
```

**Montrer le résumé les données d'entraînement**

# Performance
**Dans cette section, nous mettons les données en mémoire cache et les shuffle pour gagner en performance.**

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

# définition du model

**La bonne pratique dans un réseau neuronal étant de rendre les valeurs d'entrées petites dans ce cas RGB.
Nous allons donc normaliser les valeurs ici pour qu'elles soient entre 0 et 1**

**Le modèle se compose de trois blocs de convolution :  tf.keras.layers.Conv2D avec une couche de regroupement maximum pour chacun ( tf.keras.layers.MaxPooling2D ). 
Après cela, il ne nous reste que de transformer les données en un vecteur linéaire de dimension 1 puis les connecter à  une couche entièrement connectée ( tf.keras.layers.Dense ) activée par la fonction d'activation ReLU ( 'relu' ).**

```python
#Nombre de calsses trouvées
num_classes = len(class_names)

model = Sequential([
    #Normalisation des valeurs d'entrée
  layers.Rescaling(1./255, input_shape=(150, 150, 3)),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dropout(0.5),
  layers.Dropout(0.5,noise_shape=None, seed=None),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes)
])
```

**Show the summury of the model**

```python
model.summary()
```

# Entrainement du model V1

**Ici nous entrainons notre modèle en enregistrant le modèle en .h5, identifiant le meilleur en fonction de la precision avec save_best_only et monitor (le premiers modèles sont écrasés au fur et à mesure qu'on identifie des nouveaux meilleurs modèles)**

```python
epochs = 15
callbacks = [
    keras.callbacks.ModelCheckpoint(
    filepath="best_model.h5",
    monitor='val_accuracy',
    mode='max',
    save_best_only=True),
]
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
history = model.fit(
    train_ds, epochs=epochs, callbacks=callbacks, validation_data=val_ds,
)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np

model = ResNet50(weights='imagenet')
sunflower_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/592px-Red_sunflower.jpg"
img_path = tf.keras.utils.get_file('Red_sunflower', origin=sunflower_url)

img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0])
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```
